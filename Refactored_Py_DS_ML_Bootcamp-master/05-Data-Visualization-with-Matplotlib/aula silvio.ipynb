```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
#FUNCTIONAL
plt.plot(x,y,)
plt.xlabel('x Dudu')
plt.ylabel('y Rafa')
plt.title('Familia')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# 00
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax .plot(x,y,color='purple', lw=0.5, ls='--') 
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
